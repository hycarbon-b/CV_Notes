```python
import numpy as np
import matplotlib.pyplot as plt
# 假设有 624 个子载波信号，每个信号是长度为 N 的复数序列
num_subcarriers = 624
N = 20  # 假设每个子载波信号长度为 1024

# 生成一组随机的子载波信号（示例）
subcarriers = [np.random.randn(N) + 1j*np.random.randn(N) for _ in range(num_subcarriers)]

# 对每个子载波信号进行傅里叶逆变换
time_domain_signals = [np.fft.ifft(subcarrier) for subcarrier in subcarriers]

# 输出第一个子载波信号的时域表示
print("第一个子载波信号的时域表示：", time_domain_signals[0])
plt.plot(np.abs(time_domain_signals[0]))
```
