```python
from rdkit import Chem


# 读化合物分子·example
m = Chem.MolFromSmiles('Cc1ccccc1')   #读取Smiles


# 显示分子图片
from rdkit.Chem import Draw
Draw.MolToImage(m) # 显示分子图片
```

```python
from rdkit import Chem
from rdkit.Chem import Draw
# 用rdkit绘制20个分子的图片# List of 20 SMILES strings
smiles_list = ['Cc1ccccc1', 'CCCC', 'CO', 'C=O', 'C#N', 'CC(=O)O', 'CC(=O)N', 'CC(=O)OC', 'CC(=O)OC(=O)C', 'CC(=O)OC(=O)OC', 'CC(=O)OC(=O)OC(=O)C', 'CC(=O)OC(=O)OC(=O)OC', 'CC(=O)OC(=O)OC(=O)OC(=O)C', 'CC(=O)OC(=O)OC(=O)OC(=O)OC', 'CC(=O)OC(=O)OC(=O)OC(=O)OC(=O)C', 'CC(=O)OC(=O)OC(=O)OC(=O)OC(=O)OC', 'CC(=O)OC(=O)OC(=O)OC(=O)OC(=O)OC(=O)C', 'CC(=O)OC(=O)OC(=O)OC(=O)OC(=O)OC(=O)OC', 'CC(=O)OC(=O)OC(=O)OC(=O)OC(=O)OC(=O)OC(=O)C']

# Generate images for each molecule
for smiles in smiles_list:
    mol = Chem.MolFromSmiles(smiles)
    img = Draw.MolToImage(mol)
    img.show()
```
