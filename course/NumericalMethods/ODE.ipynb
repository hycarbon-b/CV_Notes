## 通过函数的一阶微分场图的方式计算微分方程的解  
已知一个一阶微分方程$y\prime = f(t,y),\ F(t,y)是已知的关于t,y的函数$  
步骤：  
- 绘制$(t,y,f(t,y))$的场图，z轴的数值表示该点导数（切线斜率）的大小
- 选择初始点$(t_0, y_0)$  
- $(t_1, y_1) \leftarrow (t_0+h, y_0 + f(t_0,y_0)\cdot h)$
- 不断迭代
Example:  
$$y\prime = (t-y)/2, t\in (0,5), y\in (0,4)$$   
- For $y(0)=1$, find its solution by iteration. ($y= 3e^{-t\over2}-2+t$)   
- For $y(0)=4$, find its solution by iteration. ($y= 6e^{-t\over2}-2+t$)  

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the function  
t_min, t_max, y_min, y_max = 0,5,0,4
r = 20 # resolution
t = np.linspace(0, 5, r)
y = np.linspace(0, 4, r)
t,y = np.meshgrid(t,y)
dydt = (t-y)/2
# Plot the field of derivative 
plt.quiver(t, y, np.ones_like(t), dydt, color='b')
```

```python
# Iteration  
step = 0.02
y_now = 3 #initial value
solutions = np.array([y_now])
x = np.array([t_min])
t_now = t_min
while (y_now < y_max) and (t_now < t_max): 
    t_now = t_now + step
    y_now = y_now + step * ((t_now-y_now)/2)
    x=np.append(x, t_now)
    solutions=np.append(solutions, y_now)

plt.plot(x, solutions, 'r')
# TODO: 把这个图叠加到quiver上
```
